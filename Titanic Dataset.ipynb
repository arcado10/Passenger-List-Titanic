This dataset contains the list of the passengers who had boarded the Titanic. 

The following columns/fields contains the information:
1. PassengerId: The unique id assigned to the passenger.
2. Survived: Survival Status of the Passenger, where
        1:Yes
        0:No
3. Pclass: The class of the ticket purchased.
        3:Third(Low) class ticket
        2:Second(Middle) class ticket
        1:First(High) class ticket
4. Name: The name of the passenger.
5. Sex: Sex of the passenger.
6. Age: Age of the passenger.
7. SibSp: Combined number of siblings and spouse of the passenger.
8. Parch: Combined number of parents and children of the passenger.
9. Ticket: The unique ticket Id.
10. Fare: The cost of the ticket.
11. Cabin: The cabin number booked.
12. Embarked: Port of embarkment.
        S:Southampton
        Q:Queenstown
        C:Cherbourg
